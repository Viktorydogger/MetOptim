```python
import pandas as pd
import numpy as np
import sympy as sp
from numpy import pi,cos,sin,abs,sqrt,exp
import matplotlib.pyplot as plt

in_k = pd.DataFrame({'xi':[1,2,3,4,5],'yi':[0, -1, -3, 2, -2]})

x,y,g = sp.symbols('x, y, g')


def F1(x,y):
    return sum([(x*in_k.xi[i] + y - in_k.yi[i] )**2 for i in range(5)])
def F2(x,y):
    return sum([abs(x*in_k.xi[i] + y - in_k.yi[i] ) for i in range(5)])

f1 = sum([(x*in_k.xi[i] + y - in_k.yi[i] )**2 for i in range(5)])
print( sp.diff(f1,x), sp.diff(f1,y))

beg = -10
fin = 10
```

```python
def gold_old(f,eps,a,b,X,par):
    phi = 0.5 * (1.0 + sqrt(5.0))
    dd=abs(f(b,par)-f(a,par))
    if X:
        dd=abs(f(par,b)-f(par,a))
    if dd < eps: 
        return (a+b)/2
    else:
        t = (b-a)/phi
        x1, x2 = b - t, a + t
        q=f(x1,par)
        Q=f(x2,par)
        if X:
            q=f(par,x1)
            Q=f(par,x2)
        if q>=Q:
            return gold(f, eps, x1, b,X,par)
        else:
            return gold(f, eps, a, x2,X,par)
```

```python
def gold(f,eps,a,b,X,par):
    ep = 0.000001
    c = a +(3 - sqrt(5))*(b - a)/2
    d = a +(sqrt(5) - 1)*(b - a)/2
    while(abs(a - b) >= ep):
        if X:
            ansc = f(par,c)
            ansd = f(par,d)
        else:
            ansc = f(c,par)
            ansd = f(d,par)
        if(ansc < ansd):
            b = d
            d = c
            c = a +(3 - sqrt(5))*(b - a)/2;
        else:
            a = c
            c = d
            d = a +(sqrt(5) - 1)*(b - a)/2
    return (a+b)/2
            
```

```python
def dikhotom(func):
    import pandas as pd
    import numpy as np
    ep = 10**(-6)
    de = ep/3
    a = -10
    b = 10
    while(abs(a-b)>ep):
        c = (a + b - de)/2
        d = (a + b + de)/2
        if (func.subs(g,c)<=func.subs(g,d)):
            b = d
        else:
            a = c
    x_min = (a+b)/2
    return x_min
```

```python
def coordinates(f,beg,fin):
    xmin,xmax=beg,fin
    ymin,ymax=beg,fin
    x0,y0=1,1
    eps=0.000001
    out_table = pd.DataFrame({'xn':[],'abs(delta(x))':[],'fn':[],'abs(delta(f))':[]})
    i=xo=yo=0
    while abs(f(xo,yo)-f(x0,y0))>=eps:
        i=i + 1
        xo=x0
        yo=y0
        x0=gold(f,eps,xmin,xmax,False,y0)
        y0=gold(f,eps,ymin,ymax,True,x0)
        line = {'xn':[round(x0,6),round(y0,6)],'abs(delta(x))':abs(x0 - xo),'fn':f(x0,y0),'abs(delta(f))':abs(f(x0,y0)-f(xo,yo))}
        out_table=out_table.append(line,ignore_index=True)
    print(out_table)
    return "end"
```

```python
def crushing_hops(f):
    xn = [0,0] 
    eps = 0.000001
    fdx = sp.diff(f,x)
    fdy = sp.diff(f,y)
    fz_new = f.subs([(x,xn[0]),(y,xn[1])])
    fz_old = fz_new - 1
    alf = 1
    
    out_table = pd.DataFrame({'xn':[],'f(xn)':[],'df(xn)':[]})
    
    while  abs(fz_old-fz_new) >= eps:
        gy = fdy.subs([(x,xn[0]),(y,xn[1])])
        gx = fdx.subs([(x,xn[0]),(y,xn[1])])
        x0 = xn[0] - alf*gx
        y0 = xn[1] - alf*gy
        fz_old = fz_new
        fz_new = f.subs([(x,xn[0]),(y,xn[1])])
        alf = 1
        while (fz_new - fz_old >= -1*alf*0.5*(gx**2 + gy**2)):
            alf = alf*0.5
            x0 = xn[0] - alf*gx
            y0 = xn[1] - alf*gy
            fz_new = f.subs([(x,x0),(y,y0)])   
        line = {'xn':xn,'f(xn)':round(fz_old,6),'df(xn)':round(gx**2 + gy**2,6)}
        out_table=out_table.append(line,ignore_index=True)
        xn[0] = x0
        xn[1] = y0  
    print(out_table)
    return "end"  
```
